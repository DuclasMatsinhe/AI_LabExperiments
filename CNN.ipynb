```python
# Step 1: Import libraries
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Step 2: Load dataset (Training dataset display - Requirement a)
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Show sample images
plt.figure(figsize=(6,6))
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(y_train[i])
    plt.axis('off')
plt.suptitle("Training Dataset Samples")
plt.show()

# Step 3: Normalize & reshape
X_train = X_train / 255.0
X_test = X_test / 255.0

X_train = X_train.reshape(-1,28,28,1)
X_test = X_test.reshape(-1,28,28,1)

y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)

# ------------------------------
# MODEL WITHOUT REGULARIZATION
# ------------------------------
model1 = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model1.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Requirement b: Model summary
print("\nMODEL WITHOUT REGULARIZATION:")
model1.summary()

# Train
history1 = model1.fit(X_train, y_train_cat,
                     epochs=5,
                     validation_data=(X_test, y_test_cat))

# Requirement c & d: Accuracy & Loss before regularization
plt.plot(history1.history['accuracy'], label='Train Accuracy')
plt.plot(history1.history['val_accuracy'], label='Val Accuracy')
plt.title("Accuracy (Before Regularization)")
plt.legend()
plt.show()

plt.plot(history1.history['loss'], label='Train Loss')
plt.plot(history1.history['val_loss'], label='Val Loss')
plt.title("Loss (Before Regularization)")
plt.legend()
plt.show()

# ------------------------------
# MODEL WITH REGULARIZATION (Dropout)
# ------------------------------
model2 = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])

model2.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

print("\nMODEL WITH REGULARIZATION:")
model2.summary()

# Train
history2 = model2.fit(X_train, y_train_cat,
                     epochs=5,
                     validation_data=(X_test, y_test_cat))

# Requirement e & f: Accuracy & Loss after regularization
plt.plot(history2.history['accuracy'], label='Train Accuracy')
plt.plot(history2.history['val_accuracy'], label='Val Accuracy')
plt.title("Accuracy (After Regularization)")
plt.legend()
plt.show()

plt.plot(history2.history['loss'], label='Train Loss')
plt.plot(history2.history['val_loss'], label='Val Loss')
plt.title("Loss (After Regularization)")
plt.legend()
plt.show()

# ------------------------------
# Predictions
# ------------------------------
y_pred = model2.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Requirement g: Correct predictions
correct = np.where(y_pred_classes == y_test)[0]

plt.figure(figsize=(6,6))
for i, idx in enumerate(correct[:9]):
    plt.subplot(3,3,i+1)
    plt.imshow(X_test[idx].reshape(28,28), cmap='gray')
    plt.title(f"True:{y_test[idx]} Pred:{y_pred_classes[idx]}")
    plt.axis('off')
plt.suptitle("Correct Predictions")
plt.show()

# Requirement h: Incorrect predictions
incorrect = np.where(y_pred_classes != y_test)[0]

plt.figure(figsize=(6,6))
for i, idx in enumerate(incorrect[:9]):
    plt.subplot(3,3,i+1)
    plt.imshow(X_test[idx].reshape(28,28), cmap='gray')
    plt.title(f"True:{y_test[idx]} Pred:{y_pred_classes[idx]}")
    plt.axis('off')
plt.suptitle("Incorrect Predictions")
plt.show()
```
