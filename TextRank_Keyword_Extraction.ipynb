```python
# =====================================
# STEP 1: Install Required Libraries
# =====================================
!pip install nltk networkx

import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')  # needed in newer Colab

# =====================================
# STEP 2: Import Libraries
# =====================================
import networkx as nx
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

# =====================================
# STEP 3: Input Text
# =====================================
text = """
Natural language processing enables computers to understand human language.
It is widely used in machine learning, chatbots, and speech recognition.
Keyword extraction is an important task in NLP.
"""

# =====================================
# STEP 4: Preprocessing
# =====================================
stop_words = set(stopwords.words('english'))

tokens = word_tokenize(text.lower())

# Remove stopwords and punctuation
filtered_words = [
    word for word in tokens
    if word not in stop_words and word not in string.punctuation
]

print("Filtered Words:")
print(filtered_words)

# =====================================
# STEP 5: Build Co-occurrence Graph
# =====================================
window_size = 2
graph = nx.Graph()

for i in range(len(filtered_words)):
    for j in range(i+1, i+window_size):
        if j < len(filtered_words):
            w1 = filtered_words[i]
            w2 = filtered_words[j]

            if graph.has_edge(w1, w2):
                graph[w1][w2]['weight'] += 1
            else:
                graph.add_edge(w1, w2, weight=1)

# =====================================
# STEP 6: Apply TextRank (PageRank)
# =====================================
scores = nx.pagerank(graph)

# =====================================
# STEP 7: Extract Top Keywords
# =====================================
# Sort words by score
keywords = sorted(scores.items(), key=lambda x: x[1], reverse=True)

top_n = 10
print("\nTop Keywords:")
for word, score in keywords[:top_n]:
    print(f"{word} : {score:.4f}")
```
