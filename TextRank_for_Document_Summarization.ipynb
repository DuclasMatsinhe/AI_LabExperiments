```python
# =========================================
# STEP 1: Install & Import Libraries
# =========================================
!pip install nltk networkx scikit-learn

import nltk
nltk.download('punkt')
nltk.download('punkt_tab')  # needed in newer Colab

import networkx as nx
from nltk.tokenize import sent_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# =========================================
# STEP 2: Input Document
# =========================================
text = """
Natural Language Processing (NLP) is a branch of artificial intelligence.
It helps computers understand human language.
NLP is widely used in chatbots, translation, and sentiment analysis.
Machine learning techniques improve NLP systems.
Deep learning has significantly advanced NLP applications.
"""

# =========================================
# STEP 3: Sentence Tokenization
# =========================================
sentences = sent_tokenize(text)

print("Sentences:\n")
for i, s in enumerate(sentences):
    print(f"{i}: {s}")

# =========================================
# STEP 4: Convert Sentences to TF-IDF Vectors
# =========================================
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(sentences)

# =========================================
# STEP 5: Compute Similarity Matrix
# =========================================
similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

# =========================================
# STEP 6: Build Graph & Apply PageRank
# =========================================
graph = nx.from_numpy_array(similarity_matrix)
scores = nx.pagerank(graph)

# =========================================
# STEP 7: Rank Sentences
# =========================================
ranked_sentences = sorted(
    ((scores[i], s) for i, s in enumerate(sentences)),
    reverse=True
)

# =========================================
# STEP 8: Generate Summary
# =========================================
summary_length = 2  # number of sentences in summary

summary = " ".join([ranked_sentences[i][1] for i in range(summary_length)])

print("\n🔹 Summary:\n")
print(summary)
```
