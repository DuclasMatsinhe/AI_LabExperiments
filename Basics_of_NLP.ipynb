```python
# =========================================
# STEP 1: Install & Import Libraries
# =========================================
!pip install nltk scikit-learn

import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')  # needed in newer Colab

from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import string

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# =========================================
# STEP 2: Input Text
# =========================================
text = """
Natural Language Processing (NLP) is a field of Artificial Intelligence.
It helps computers understand human language.
NLP is used in chatbots, translation, and sentiment analysis.
"""

# =========================================
# STEP 3: Sentence Tokenization
# =========================================
sentences = sent_tokenize(text)
print("Sentences:\n", sentences)

# =========================================
# STEP 4: Word Tokenization
# =========================================
words = word_tokenize(text.lower())
print("\nWords:\n", words)

# =========================================
# STEP 5: Remove Stopwords & Punctuation
# =========================================
stop_words = set(stopwords.words('english'))

filtered_words = [
    word for word in words
    if word not in stop_words and word not in string.punctuation
]

print("\nFiltered Words:\n", filtered_words)

# =========================================
# STEP 6: Stemming
# =========================================
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]

print("\nStemmed Words:\n", stemmed_words)

# =========================================
# STEP 7: Lemmatization
# =========================================
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

print("\nLemmatized Words:\n", lemmatized_words)

# =========================================
# STEP 8: Bag of Words (BoW)
# =========================================
corpus = [" ".join(filtered_words)]

bow = CountVectorizer()
X_bow = bow.fit_transform(corpus)

print("\nBoW Features:\n", bow.get_feature_names_out())
print("BoW Vector:\n", X_bow.toarray())

# =========================================
# STEP 9: TF-IDF
# =========================================
tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(corpus)

print("\nTF-IDF Features:\n", tfidf.get_feature_names_out())
print("TF-IDF Vector:\n", X_tfidf.toarray())

# =========================================
# STEP 10: N-Grams (Bigrams)
# =========================================
bigram_vectorizer = CountVectorizer(ngram_range=(2,2))
X_bigrams = bigram_vectorizer.fit_transform(corpus)

print("\nBigrams:\n", bigram_vectorizer.get_feature_names_out())
print("Bigram Vectors:\n", X_bigrams.toarray())
```
