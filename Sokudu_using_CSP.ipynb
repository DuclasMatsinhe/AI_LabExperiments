```python
# =========================================
# Sudoku Solver using CSP (Backtracking + MRV)
# =========================================

# Example Sudoku (0 = empty cell)
board = [
    [5,3,0,0,7,0,0,0,0],
    [6,0,0,1,9,5,0,0,0],
    [0,9,8,0,0,0,0,6,0],
    [8,0,0,0,6,0,0,0,3],
    [4,0,0,8,0,3,0,0,1],
    [7,0,0,0,2,0,0,0,6],
    [0,6,0,0,0,0,2,8,0],
    [0,0,0,4,1,9,0,0,5],
    [0,0,0,0,8,0,0,7,9]
]

# =========================================
# Print Sudoku Board
# =========================================
def print_board(b):
    for i in range(9):
        if i % 3 == 0 and i != 0:
            print("-"*21)
        for j in range(9):
            if j % 3 == 0 and j != 0:
                print("|", end=" ")
            print(b[i][j] if b[i][j] != 0 else ".", end=" ")
        print()

# =========================================
# Check if valid assignment
# =========================================
def is_valid(b, row, col, num):
    # Row
    if num in b[row]:
        return False

    # Column
    for i in range(9):
        if b[i][col] == num:
            return False

    # 3x3 Box
    start_row = row - row % 3
    start_col = col - col % 3

    for i in range(3):
        for j in range(3):
            if b[start_row+i][start_col+j] == num:
                return False

    return True

# =========================================
# Find Empty Cell with MRV Heuristic
# =========================================
def find_mrv(b):
    min_options = 10
    best_cell = None

    for i in range(9):
        for j in range(9):
            if b[i][j] == 0:
                options = 0
                for num in range(1, 10):
                    if is_valid(b, i, j, num):
                        options += 1

                if options < min_options:
                    min_options = options
                    best_cell = (i, j)

    return best_cell

# =========================================
# Backtracking Solver
# =========================================
def solve(b):
    cell = find_mrv(b)

    if not cell:
        return True  # solved

    row, col = cell

    for num in range(1, 10):
        if is_valid(b, row, col, num):
            b[row][col] = num

            if solve(b):
                return True

            b[row][col] = 0  # backtrack

    return False

# =========================================
# Run Solver
# =========================================
print("Original Sudoku:\n")
print_board(board)

if solve(board):
    print("\nSolved Sudoku:\n")
    print_board(board)
else:
    print("No solution found")
```
