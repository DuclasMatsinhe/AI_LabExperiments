```python
# ================================
# STEP 1: Install & Import
# ================================
!pip install nltk gensim

import nltk
nltk.download('punkt')
nltk.download('punkt_tab')  # important for new Colab versions

from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec

# ================================
# STEP 2: Input Text Data
# ================================
text = """
Natural language processing is very interesting.
Machine learning makes NLP powerful.
Word embeddings capture meaning of words.
"""

# ================================
# STEP 3: Preprocessing
# ================================
# Convert to lowercase and tokenize sentences
sentences = text.lower().split('\n')

# Tokenize each sentence into words
data = [word_tokenize(sentence) for sentence in sentences if sentence.strip() != ""]

print("Tokenized Sentences:")
print(data)

# ================================
# STEP 4: Train Word2Vec Model
# ================================
model = Word2Vec(
    sentences=data,
    vector_size=100,   # size of embedding vector
    window=5,          # context window
    min_count=1,       # include all words
    workers=4
)

# ================================
# STEP 5: Word Embeddings
# ================================
print("\nVector for word 'language':")
print(model.wv['language'])

# ================================
# STEP 6: Similar Words
# ================================
print("\nWords similar to 'language':")
print(model.wv.most_similar('language'))

# ================================
# STEP 7: Vocabulary
# ================================
print("\nVocabulary:")
print(list(model.wv.index_to_key))

# ================================
# STEP 8: Save Model (Optional)
# ================================
model.save("word2vec.model")

# ================================
# STEP 9: Load Model (Optional)
# ================================
# from gensim.models import Word2Vec
# loaded_model = Word2Vec.load("word2vec.model")
```
