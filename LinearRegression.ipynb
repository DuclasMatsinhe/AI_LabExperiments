```python
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Dataset (you can replace with your own)
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Step 3: Plot ONLY datapoints (Requirement a)
plt.scatter(X, y)
plt.title("Datapoints Scattering (Without Best Fit Line)")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Step 4: Initialize parameters
m = 0
c = 0
learning_rate = 0.01
epochs = 1000
n = len(X)

# Step 5: Gradient Descent Training
for i in range(epochs):
    y_pred = m * X + c

    # Compute gradients
    dm = (1/n) * sum((y_pred - y) * X)
    dc = (1/n) * sum(y_pred - y)

    # Update parameters
    m = m - learning_rate * dm
    c = c - learning_rate * dc

# Step 6: Plot datapoints + best fit line (Requirement b)
plt.scatter(X, y, label="Data Points")
plt.plot(X, m*X + c, color='red', label="Best Fit Line")
plt.title("Scatter Plot with Best Fit Line (Final Iteration)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
